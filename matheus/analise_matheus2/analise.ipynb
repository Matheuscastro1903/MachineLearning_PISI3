# Análise de Correlação: Existe um padrão de gasto com transporte?

Antes de segmentarmos por geografia, precisamos entender o comportamento humano básico diante do consumo. A hipótese que testaremos aqui é de que o custo com transporte é ditado primariamente pelo nível de **renda**.

### O Fenômeno da "Acomodação de Estilo de Vida"
Embora o custo fixo de combustíveis ou passagens varie pouco, os dados sugerem que, à medida que a renda sobe, o indivíduo tende a trocar a **eficiência barata** (transporte público) pela **conveniência custosa** (veículos próprios ou aplicativos). Este comportamento ocorre de forma similar tanto em metrópoles quanto no interior.

---
## Análise do gráfico

O gráfico de dispersão abaixo projeta 20.000 registros para verificar se existe uma ligação direta entre o ganho e o gasto. A visualização dos dados evidencia que o gasto com transporte acompanha o crescimento da renda. A linha de tendência reforça este padrão, indicando que o custo de locomoção escala proporcionalmente ao aumento do poder aquisitivo do indivíduo.


---

## O que é a Linha de Tendência?

A linha vermelha é um modelo matemático de **Regressão Linear** conhecido como **OLS** (*Ordinary Least Squares*). Ela não é apenas um traçado visual, mas o resultado de um processo rigoroso de ajuste:



### Como o cálculo é realizado:

1. **Cálculo dos Resíduos:** O algoritmo mede a distância vertical de cada ponto azul (dado real) até a reta (calculada anteriormente).
2. **Penalização por Quadrados:** Essas distâncias são elevadas ao quadrado. Isso evita que erros negativos e positivos se anulem e dá mais peso aos pontos que estão muito distantes da média.
3. **Minimização Global:** O computador testa infinitas posições para a reta até encontrar aquela onde a **soma de todos esses quadrados é a mínima possível**.


A inclinação positiva da reta ($\beta_1 > 0$) prova que o transporte é um **gasto elástico**(aquele que muda significativamente de acordo com a variação da sua renda ou do preço do produto.). 



```python
import plotly.express as px

# Criando o gráfico de dispersão para provar a correlação Renda vs. Gasto
fig = px.scatter(
    dados,                          #dataset
    x="Income",                     #eixo x:renda
    y="Transport",                  #eixo y:transporte
    
    
    #trend line desenha uma linha de tendência
    #construido a partir da equação y=ax+b(equação afim)
    trendline="ols",                
    trendline_color_override="red", #apenas para estilizar a linha de tendência
    
    title="Correlação Direta: Crescimento da Renda vs. Gasto com Transporte",
    labels={
        "Income": "Renda Mensal ($)",
        "Transport": "Custo de Transporte ($)"
    },
    
    template="plotly_white",
    opacity=0.4                     
)

#apenas estilizando os pontos para melhorar a visualização
fig.update_traces(marker=dict(size=6, color='royalblue'))

fig.show()
```

Análise Comparativa: O Impacto do City Tier

Esta análise visa confrontar o senso comum de que a localização geográfica é o principal determinante dos custos de vida. Nossa hipótese é que os gastos com transporte nos **Tiers 1, 2 e 3** são estatisticamente semelhantes, sugerindo que o real motor do gasto não é o CEP, mas sim a **Renda disponível**.

### Evidências Visuais: Dispersão e Concentração
Ao observarmos o **Gráfico de Pontos (Scatter Plot)** e o **Gráfico de Caixa (Boxplot)** em conjunto, notamos um comportamento homogêneo:
* **Distribuição:** A nuvem de pontos revela que a dispersão dos gastos segue o mesmo padrão em todas as cidades, com *outliers* (pontos fora da curva) presentes em todos os níveis.
* **Equilíbrio:** O Boxplot confirma que as medianas são bem semelhantes,não variando drasticamente entre as cidades.

---

### Fundamentação Técnica: Por que usar o Boxplot?

O **Boxplot (Gráfico de Caixa)** é uma ferramenta de análise de dispersão que organiza os dados em ordem crescente para dividi-los em **quartis** (quatro partes de 25% cada). Diferente de uma média aritmética simples, ele oferece uma visão profunda da estrutura dos dados:



#### **As fatias da população:**
* **25% (Q1):** Representa a faixa de menor gasto.
* **50% (Mediana):** O valor exato que divide a população ao meio, indicando o comportamento do "consumidor típico".
* **75% (Q3):** Define o "teto" de gastos para a grande maioria da base de dados.

> **O Valor Analítico:** Este gráfico nos permite afirmar com precisão que **"X% da base de dados gasta até o valor Y"**. Ao destacar a mediana e os quartis, o Boxplot neutraliza o efeito de valores extremos (*outliers*), fornecendo uma métrica realista e robusta para o planejamento financeiro.




```python
import plotly.express as px

#criando um gráfico de caixa
fig = px.box(
    dados,                          #dataset fonte
    x="City_Tier",                  #Eixo X
    y="Transport",                  #Eixo Y
    color="City_Tier",              # 4. Diferenciação: Aplica uma cor para cada Tier facilitando a leitura
    
    # 5. Mapeamento de Cores: Definimos manualmente as cores para cada categoria
    # No Boxplot, o preenchimento da caixa assume essas cores de forma sólida e opaca
    color_discrete_map={
        "Tier_1": "royalblue",
        "Tier_2": "goldenrod",
        "Tier_3": "indianred"
    },
    
    # Define quais pontos mostrar: "outliers" limpa o gráfico e foca nos casos extremos
    points="all",              
    
    title="Distribuição do Gasto com Transporte por Tier de Cidade", #título do gráfico
    labels={                        #apenas renomeando os eixos 
        "City_Tier": "Nível da Cidade",
        "Transport": "Custo de Transporte ($)"
    },
    
    template="plotly_white"         # 7. Estética: Um fundo limpo para destacar os dados
)

# 8. AJUSTE DE OPACIDADE: Força a "caixa sólida" a ter 100% de cor (sem transparência)
fig.update_traces(marker_opacity=1.0)

# O comando 'show()' renderiza o gráfico interativo na sua tela
fig.show()
```

A análise deste gráfico de dispersão segmentado por **City Tier** é a peça final que corrobora nossa tese inicial. Aqui, isolamos o impacto da renda dentro de cada nível de cidade para verificar se a localização geográfica alteraria a inclinação do consumo.

# Objetivo da Análise

O objetivo desta análise é demonstrar que variáveis como o número de dependentes, profissão e idade não são fundamentais para definir se o usuário terá uma boa gestão do seu dinheiro.

---

## Metodologia

Analisaremos essas variáveis em relação ao **Savings Ratio** (Taxa de Economia), que busca identificar quanto da renda está sendo "perdido" em gastos supérfluos.

---

## O que o seu "Percentual" realmente significa?

O cálculo soma gastos com lazer, alimentação fora e outros itens não essenciais. Portanto:

- **Percentual ALTO (Ex: 40%):** Indica que a pessoa é **ineficiente financeiramente**. Ela gasta quase metade do que ganha com itens não essenciais. Existe um enorme "potencial" de economia sendo desperdiçado.

- **Percentual BAIXO (Ex: 5%):** Indica que a pessoa é **eficiente**. O orçamento está tão ajustado que quase não sobra margem para cortar gastos supérfluos.

---

## 4.1. Análise: Potencial de Economia vs. Dependentes
A análise indica que pessoas sem dependentes tendem a apresentar melhores índices de potencial de economia. No entanto, ao observar a densidade, notamos que a grande massa de dados se concentra de forma muito semelhante entre os grupos.

> Utilizamos o método `update_traces(scalemode='count')` com o objetivo de deixar o violino com o tamanho real da quantidade de dados e, mesmo assim, podemos observar que a **"grande massa"** das pessoas que têm **0 dependentes** e das pessoas que têm **4 dependentes** está concentrada mais ou menos na mesma área.

**O número de dependentes influencia o orçamento, mas não é o fator fundamental**. A concentração de pessoas com 0 ou 4 dependentes permanece majoritariamente na mesma faixa percentual de gastos.


```python
import plotly.express as px

#lista vai servir para somar os valores
colunas_economia = [
    'Potential_Savings_Groceries', 'Potential_Savings_Transport', 
    'Potential_Savings_Eating_Out', 'Potential_Savings_Entertainment', 
    'Potential_Savings_Utilities', 'Potential_Savings_Healthcare', 
    'Potential_Savings_Education', 'Potential_Savings_Miscellaneous'
]
#criando nova coluna em relação aos gastos
dados['Potential_Savings'] = dados[colunas_economia].sum(axis=1)
#calculando o índice de 'gasto desnecessário'
dados['waste_Ratio'] = (dados['Potential_Savings'] / dados['Income']) * 100

#necessário passar apra string para que o plotly divida em categorias

dados['Dependents_Str'] = dados['Dependents'].astype(str)

# 3. Gerando o Violin Plot por Dependentes
fig = px.violin(
    dados, 
    x="Dependents_Str", 
    y="waste_Ratio",  
    # color="Dependents_Str",  <-- REMOVA ESTA LINHA
    color_discrete_sequence=['#7f7f7f'],
    box=True,           
    points="all",      
    category_orders={"Dependents_Str": sorted(dados['Dependents_Str'].unique())},
    title="Distribuição de Potencial Economia por Número de Dependentes",
    labels={
        "Dependents_Str": "Número de Dependentes",
        "waste_Ratio": "% de Economia sobre a Renda"
    },
    template="plotly_white" # O fundo branco já ajuda no visual clean
)
# linha da média para comparar com a mediana (boxplot interno)
fig.update_traces(meanline_visible=True)

#esse método é responsávell por deixar o violino no tamanho real,sem haver distorções na imagem,mesmo se os outros
#forem maiores
fig.update_traces(scalemode='count')

fig.update_traces(
    opacity=0.6,          #deixa o preenchimento suave
    line_width=1.5,       #engrossa um pouco a borda do violino
    line_color='black',   #força a borda a ser preta/escura para contraste
    meanline_visible=True,
    meanline_width=3,     #destaca a linha da média
)
fig.update_layout(
    hovermode="x", # Faz a linha focar no grupo (eixo X) onde o mouse está
    yaxis=dict(
        showspikes=True, #ativa a linha de guia
        spikemode="across", #faz a linha atravessar o gráfico inteiro
        spikesnap="cursor", #a linha segue exatamente o mouse, não "pula" nos pontos
        spikethickness=2, #espessura da linha
        spikedash="dash", #estilo da linha 
        spikecolor="gray" 
    )
)


fig.show()
```

## 4.2. Análise: Potencial de Economia x Idade
O comportamento por faixa etária assemelha-se ao de dependentes. Embora o grupo de idosos possua uma amostra significativamente maior no dataset, a densidade do comportamento financeiro (onde o violino é mais "gordo") é muito parecida com as demais idades.

Mesmo com o avanço da idade, a **densidade de gastos supérfluos mantém um padrão, reforçando que a idade não determina a maturidade financeira no consumo.**


```python
import pandas as pd
import plotly.express as px

#preparação dos Dados (Mantendo suas variáveis)
colunas_economia = [
    'Potential_Savings_Groceries', 'Potential_Savings_Transport', 
    'Potential_Savings_Eating_Out', 'Potential_Savings_Entertainment', 
    'Potential_Savings_Utilities', 'Potential_Savings_Healthcare', 
    'Potential_Savings_Education', 'Potential_Savings_Miscellaneous'
]

dados['Potential_Savings'] = dados[colunas_economia].sum(axis=1)
dados['waste_Ratio'] = (dados['Potential_Savings'] / dados['Income']) * 100

#criando a faixa etária
bins = [0, 18, 31, 46, 61, 100]
labels_idade = ['Jovem', 'Jovem Adulto', 'Adulto', 'Sênior', 'Idoso']
dados['Age_Group'] = pd.cut(dados['Age'], bins=bins, labels=labels_idade, right=False)

# Gerando o Violin Plot com o Estilo Padrão
fig = px.violin(
    dados, 
    x="Age_Group", 
    y="waste_Ratio", 

    color_discrete_sequence=['#7f7f7f'], 
    box=True,            
    points='all',         
    category_orders={"Age_Group": labels_idade},
    title="Distribuição de Potencial Economia por Faixa Etária",
    labels={
        "Age_Group": "Faixa Etária",
        "waste_Ratio": "% de Economia sobre a Renda"
    },
    template="plotly_white"
)


fig.update_traces(
    scalemode='count',    #mantém o tamanho real da amostra
    opacity=0.6,          #preenchimento 
    line_width=1.5,       #contorno definido
    line_color='black',   #contorno em preto para contraste
    meanline_visible=True,
    meanline_width=3      #média em destaque
)


fig.update_layout(
    hovermode="x", 
    yaxis=dict(
        showspikes=True, 
        spikemode="across", 
        spikesnap="cursor", 
        spikethickness=2, 
        spikedash="dash", 
        spikecolor="gray" 
    )
)

fig.show()
```



## 4.3. Análise: Potencial de Economia x Ocupação
Nesta última análise, verificamos a relação entre a profissão e a gestão da renda. O padrão observado anteriormente se repete: a maior densidade de dados converge para a mesma porcentagem, independentemente do cargo ocupado.

**A ocupação influencia o volume de renda, mas não é o fator determinante para a eficiência do gasto.**

```python
import plotly.express as px


colunas_economia = [
    'Potential_Savings_Groceries', 'Potential_Savings_Transport', 
    'Potential_Savings_Eating_Out', 'Potential_Savings_Entertainment', 
    'Potential_Savings_Utilities', 'Potential_Savings_Healthcare', 
    'Potential_Savings_Education', 'Potential_Savings_Miscellaneous'
]
dados['Potential_Savings'] = dados[colunas_economia].sum(axis=1)
dados['waste_Ratio'] = (dados['Potential_Savings'] / dados['Income']) * 100

ordem_ocupacao = ['Estudante', 'CLT', 'Autônomo', 'Aposentado']

fig = px.violin(
    dados, 
    x="Occupation",      
    y="waste_Ratio", 
    
    color_discrete_sequence=['#7f7f7f'], 
    box=True,          
    points="all",      
    category_orders={"Occupation": ordem_ocupacao}, 
    title="Análise de Potencial de Economia: Distribuição por Ocupação",
    labels={
        "Occupation": "Tipo de Ocupação",
        "waste_Ratio": "% de Economia Potencial"
    },
    template="plotly_white"
)


fig.update_traces(
    scalemode='count',    
    opacity=0.6,          
    line_width=1.5,       
    line_color='black',   
    meanline_visible=True,
    meanline_width=3      
)


fig.update_layout(
    hovermode="x", 
    yaxis=dict(
        showspikes=True, 
        spikemode="across", 
        spikesnap="cursor", 
        spikethickness=2, 
        spikedash="dash", 
        spikecolor="gray" 
    )
)

fig.show()
```


## 5. Conclusão Geral
Dessa forma, concluímos que variáveis como idade, ocupação e número de dependentes exercem influência sobre o orçamento, mas não são fatores determinantes para a eficiência financeira.

O verdadeiro fator decisivo parece ser o **nível de educação financeira**. A falta de uma base educacional sólida sobre finanças gera um padrão de consumo comum entre diferentes grupos sociais. Isso faz com que indivíduos de perfis completamente distintos acabem apresentando o mesmo comportamento de gastos ineficientes, criando um "padrão social" de desperdício financeiro.
